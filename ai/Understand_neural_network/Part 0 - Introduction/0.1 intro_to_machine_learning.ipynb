## **1.1. General Introduction to Machine Learning**

### **1.1.1. Definition and Role of Machine Learning in Artificial Intelligence**

**Machine learning** is a subfield of artificial intelligence (AI). In simple terms, it's a method that enables a computer to learn from data without being explicitly programmed for each task.

Imagine you want to teach a computer to recognize cats in photos. Traditionally, you would have written rules to help it understand what a cat is. With machine learning, instead of giving precise instructions, you provide the computer with lots of images of cats and other objects. It learns to distinguish cats by finding patterns or characteristics in these images.

### **1.1.2. Types of machine learning**

Machine learning has three principal types:

### Supervised Learning
- **Description**: The model is trained on labeled data, where each input has a known output.
- **Goal**: Learn the relationship between inputs and outputs to predict new data.
- **Examples**: Spam detection, image classification, and medical diagnosis.

### Unsupervised Learning
- **Description**: The model is trained on data without labels to find hidden patterns or structures.
- **Goal**: Discover groups or patterns in the data.
- **Examples**: Customer segmentation, anomaly detection, and recommendation systems.

### Reinforcement Learning
- **Description**: An agent learns by interacting with an environment, receiving rewards or penalties.
- **Goal**: Maximize cumulative rewards over time through trial and error.
- **Examples**: Robotics, game playing, and autonomous vehicles.

---

### **1.1.3. Important elements of machine learning**

The machine learning process relies on three key elements:

1. **Data**:
    
    Data is the fuel for machine learning. The more data there is, the better (as long as it is of good quality). A model will learn more effectively with a large number of varied examples.
    
2. **The Model**:
    
    A model is essentially a mathematical function that transforms input data (the features) into predictions or decisions. For example, a model can learn to say "this is a cat" or "this is not a cat" based on the features of an image.
    
3. **Algorithms**:
    
    These are the methods or techniques used to adjust and refine the model so that it learns correctly from the data. Algorithms are responsible for finding patterns or relationships in the data.

## **1.2. Transition to Neural Networks**

A neural network is an algorithmic method that allows a model to learn from data by adjusting its "neurons" (or nodes) to identify patterns within the data.

### **1.2.1. Neural Networks as a Type of Algorithm:**

An **algorithm** in machine learning is a series of mathematical steps or instructions that enable a model to learn from data. In the case of neural networks, the algorithm operates as follows:

1. **Take input data** (for example, an image, text, or numbers).
2. **Pass it through multiple layers of neurons**, where each layer performs specific calculations to detect certain patterns or features in the data.
3. **Adjust the weights** (or connections between neurons) based on the errors made by the model during prediction. This process is known as **learning** or **training**.
4. Repeat this process until the neural network is capable of making accurate predictions.

---

### **1.3 Overview**

Here is a summary of the essential parts of a neural network that we will be covering today:

**Neurons (Nodes)**
   - Basic units that process inputs and generate outputs.

**Layers**
   - **Input Layer**: Receives initial data.
   - **Hidden Layers**: Process data through multiple transformations.
   - **Output Layer**: Produces the final prediction or output.

**Weights**
   - Parameters that determine the influence of inputs on outputs.

**Bias**
   - Additional parameter that allows neurons to activate even with zero inputs.

**Activation Functions**
   - Introduce non-linearity, enabling the network to learn complex patterns (e.g., ReLU, Sigmoid).

**Forward Propagation**
   - Process of passing input data through the network to generate output.

**Loss Function**
   - Measures the error between predicted and actual values, guiding learning.

**Backpropagation**
   - Propagate the error backwards in the neural network.

**Optimizer**
   - Algorithm that adjusts weights and biases to minimize the loss (e.g., SGD, Adam).

**Learning Rate**
   - Controls the size of updates to the weights during training.

These components work together to allow a neural network to learn from data and make predictions.
